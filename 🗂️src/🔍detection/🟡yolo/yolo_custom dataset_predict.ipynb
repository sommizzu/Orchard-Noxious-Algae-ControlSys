```python
from ultralytics import YOLO
import cv2
import os
import csv

# Define the directory for images
image_dir = "/content/drive/MyDrive/images"

# Define the output directory
output_dir = "/content/drive/MyDrive/yolov8n_output"
os.makedirs(output_dir, exist_ok=True)

# Load YOLOv8 model
model = YOLO('yolov8n.pt')  # Pretrained YOLOv8n model

# Define the CSV file path
csv_path = os.path.join(output_dir, "yolov8_predictions.csv")
with open(csv_path, 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(["image_path", "object_count", "label", "confidence", "x1", "y1", "x2", "y2"])

# List of image files
image_files = [file for file in os.listdir(image_dir) if file.lower().endswith(('.jpg', '.jpeg', '.png'))]

# Perform predictions for each image
results = model([os.path.join(image_dir, file) for file in image_files])

for result, image_file in zip(results, image_files):
    # Display results
    result.show()  # Optional, for visual inspection

    # Save annotated image
    annotated_image_path = os.path.join(output_dir, f"{image_file}_annotated.jpg")
    result.save(filename=annotated_image_path)

    # Write bounding boxes to CSV
    with open(csv_path, 'a', newline='') as csvfile:
        csv_writer = csv.writer(csvfile)
        for (x1, y1, x2, y2, conf, cls) in result.boxes.data:
            csv_writer.writerow([os.path.join(image_dir, image_file), len(result.boxes.data), result.names[int(cls)], conf, x1, y1, x2, y2])
```
